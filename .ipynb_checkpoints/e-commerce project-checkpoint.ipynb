```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # for plotting graphs
import seaborn as sns # for plotting graphs
import datetime as dt

#Fonction qui va nous permettre de récupérer les données fichiers .csv
def load_data(name):
    file_path = os.path.join("D:\Projet e-commerce", name)
    return pd.read_csv(file_path)

#on récupère l'ensemble des datasets
customers = load_data("olist_customers_dataset.csv")
geolocalisation = load_data("olist_geolocation_dataset.csv")
order_items = load_data("olist_order_items_dataset.csv")
order_payments = load_data("olist_order_payments_dataset.csv")
order_reviews = load_data("olist_order_reviews_dataset.csv")#
orders = load_data("olist_orders_dataset.csv")
products = load_data("olist_products_dataset.csv")
sellers = load_data("olist_sellers_dataset.csv")
product_cnt = load_data("product_category_name_translation.csv")
customers
```

```python
#fusion des dataframes : à la fin, il nous reste 2 dataframes : Data et geolocalisation
Product = products.join(product_cnt.set_index('product_category_name'), on = 'product_category_name')
order1 = orders.join(order_payments.set_index('order_id'), on = 'order_id')
order2 = order1.join(order_reviews.set_index('order_id'), on = 'order_id')
Order = order2.join(order_items.set_index('order_id'), on = 'order_id')
Order_sellers = Order.join(sellers.set_index('seller_id'), on = 'seller_id')
Orders = Order_sellers.join(Product.set_index('product_id'), on = 'product_id')
Data = Orders.join(customers.set_index('customer_id'), on = 'customer_id')
Data
```

```python
#commencons la segmentation des données 
Data.customer_city.value_counts()[:10].plot(kind='bar')
```

```python
#Etude démographhique pour la ville de Sao Paulo
Data=Data[Data.customer_city=='belo horizonte']
Data['order_purchase_timestamp'].min(),Data['order_purchase_timestamp'].max()
#date actuelle
PRESENT = dt.datetime(2018,10,18)
#change la colonne "order_purchase_timestamp" en datetime
Data['order_purchase_timestamp'] = pd.to_datetime(Data['order_purchase_timestamp'])
Data
```

```python
#on insère une colonne pour remplacer le 'customer_id'

#on récupère les colonnes qui vont nous permettre de déterminer la Recence, la Fréquence et le Montant
Data=Data[['customer_id','order_purchase_timestamp', 'order_id', 'customer_zip_code_prefix','price', 'product_photos_qty']]
Data['TotalPrice'] = Data['product_photos_qty'] * Data['price']

#création de la table rfm (récence, fréquence, montante)
Data['Dayssinceorder'] = Data['order_purchase_timestamp'].apply(lambda x: (PRESENT - x).days)
aggr = {
    'Dayssinceorder': lambda x: x.min(),  # the number of days since last order (Recency)
    'order_id': lambda num: len(num), # the total number of orders in the last period (Frequency)
    'TotalPrice': lambda price: price.sum(),
}
rfm = Data.groupby('customer_id').agg(aggr).reset_index()
rfm.rename(columns={'Dayssinceorder': 'Recency', 'order_id': 'Frequency', 'TotalPrice': 'Monetary'}, inplace=True)
rfm
```

```python
#Recency graph for the dataset
plt.figure(figsize=(8,5))
sns.distplot(rfm.Recency, kde=False, rug=True)
```

```python
#nous allons maintenant calculer les R, F et M scores
quintiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([.2, .6]).to_dict()
quintiles

def r_score(x):
    if x <= quintiles['Recency'][.2]:
        return 3
    elif x <= quintiles['Recency'][.6]:
        return 2
    else:
        return 1

def fm_score(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.6]:
        return 2
    else:
        return 3   

rfm['R'] = rfm['Recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
rfm['M'] = rfm['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))
#on concatène le tout
rfm['RFM_score'] = rfm['R'].astype(str)+rfm['F'].astype(str)+rfm['M'].astype(str)
rfm
```

```python
rfm = rfm.astype({"RFM_score": int})
rfm.info()
```

```python
#création de la colonne level pour classer les clients en fonstion de leur score
rfm['RFM_level'] = ""
best = rfm.loc[rfm['RFM_score'] >= 311].index
lost_cheap = rfm.loc[rfm['RFM_score'] <= 113].index
lost = rfm.loc[(rfm['RFM_score'] > 113) & (rfm['RFM_score'] <= 133)].index
attention = rfm.loc[(rfm['RFM_score'] >= 211) & (rfm['RFM_score'] < 221)].index
others = rfm.loc[(rfm['RFM_score'] >= 221) & (rfm['RFM_score'] < 311)].index
                
for i in rfm.index:
    if i in lost_cheap:
        rfm.RFM_level.iloc[i]='Lost Cheap Customers'
    elif i in lost:
        rfm.RFM_level.iloc[i]='Lost Customers'
    elif i in best:
        rfm.RFM_level.iloc[i]='Best Customers'
    elif i in attention:
        rfm.RFM_level.iloc[i]='Need attention'
    elif i in others:
        rfm.RFM_level.iloc[i]='Normal Customers'
    else:
        rfm.RFM_level.iloc[i]='Others'

rfm
```

```python
sq1 = rfm.groupby('RFM_level')['customer_id'].nunique().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14,8))
sns.barplot(data=sq1, x='RFM_level', y='customer_id', palette='Greens_d', orient=True)
```

```python
# Calculate average values for each RFM_Level, and return a size of each segment 
rfm_level_agg = rfm.groupby('RFM_level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)
# Print the aggregated dataset
print(rfm_level_agg)
```

```python
#schéma pour visualiser les segments de clients
rfm_level_agg.columns = ['RecencyMean','FrequencyMean','MonetaryMean', 'Count']
#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 9)
plt.pie(x=sq1.customer_id, labels=sq1.RFM_level, autopct="%.1f%%")
plt.title("RFM Segments",fontsize=18,fontweight="bold")
plt.axis('off')
plt.show()
```

```python
# count the number of customers in each segment
segments_counts = rfm['RFM_level'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['Lost Cheap Customers', 'Need attention']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```
